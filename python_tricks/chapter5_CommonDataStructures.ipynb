## 5.1 Dictionaries, Maps and Hashtables

- Dictionaries (dict) stores an arbitrary number of objects, each identified by a unique dictionary _key_.
- Dictionaries are often called _maps_, _hashmaps_, _lookup tables_, or _associative arrays_. They allow for efficient lookup, insertion, and deletion of any object associated with a given key.
- The keys of a dictionary should be __hashable__, which means it has a hash value which never changes during its lifetime. All the immutable objects are hashable, but not all the hashable objects are immutable (eg: create a class with \__hash__ method).
- Dictionaries are highly optimized and underlie many parts of teh language, it has O(1) time complexity for lookup, insert, update, and delete operations in the average case.

Specialized dictionaries:
- _collections.OrderedDict_: remember the insertion order of keys
- _collections.defaultdict_: return default values for missing keys
- _collections.ChainMap_: search multiple dictionaries as a single mapping
- _types.MappingProxyType_: a wrapper for making read-only dictionaries

```python
# base dictionary examples

phonebook = {
    'Alice': 12434,
    'Gaby': 50342,
    'Toby': 92143,
}
```

```python
squares = {x:x**2 for x in range(5)}
squares
```

```python
# Specialized dictionaries: OrderedDict

from collections import OrderedDict

dd1 = OrderedDict(trade1 = 'buy', trade3 = 'sell')
dd2 = {'trade1': 'buy', 'trade3': 'sell'}
```

```python
dd1
```

```python
dd1['trade2'] = 'buy'
dd2['trade2'] = 'buy'
```

```python
dd1
```

```python
dd2
```

```python
dd1.keys()
```

```python
# specialized dictionaries: collections.defaultdict
from collections import defaultdict

dd = defaultdict(list)
```

```python
dd
```

```python
dd['dogs'] .append('Rufus')
```

```python
dd['dogs'].append('Jimmy')
```

```python
dd
```

```python
dd['dogs']
```

```python
dd = dict({})
print(dd)
dd['dogs'].append('Jimmy')
```

```python
dd = defaultdict(list)
print(dd)
dd['dogs'].append('Jimmy')
```

```python
dd
```

```python
# specialized dictionaries: collections.ChainMap

from collections import ChainMap
dict1 = {'first':1, 'second':2}
dict2 = {'third':3, 'fourth': 4}
chain = ChainMap(dict1, dict2)
```

```python
chain
```

```python
chain['third']
```

```python
dict3 = dict1, dict2
```

```python
dict3
```

```python
dict3['third']
```

```python
dict2.update(dict1)
dict2
```

```python
dict2['third']
```

```python
# Specialized dictionaries: types.MappingProxyType

from types import MappingProxyType

writable = {'one':1, 'two':2}
read_only = MappingProxyType(writable)
```

```python
writable['three'] = 3
writable
```

```python
read_only['three']=3
```
