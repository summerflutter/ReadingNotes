## 4.1 Object Comparison: "is" VS "=="

- "is": compares identities. An __is__ expression evaluates to True if two variables point to the same (identical) object.
- "==": compares equalities. An __==__ expression evaluates to True if the objects referred to by the variables are equal (have the same contents). 

```python
a = [1,2,3]
b = a 
```

```python
a is b
```

```python
a == b
```

```python
c = list(a)
c
```

```python
c == a
```

```python
c is a
```

## 4.2 String Conversion: Every Class Needs a \__repr__ 

2 dunder methods:
- \__str__: the dunder method which gets called when we try to convert an object into a string through the various means that are available, for example _print()_ and _str()_ functions.
- \__repr__: the dunder method used to inspect an object in a Python interpreter session, for example inspecting the class directly.

Difference between these 2 methods:
- \_str_: the output should primarily be __readable__. It's meant to return a concise textual representation for human consumption - something you'd feel comfortable displaying to a user..  
- \_repr_: the output should be __unambiguous__. The resulting string is intended more as a debugging aid for developers, so it needs to be as explicit as possble about what this object is.
- Containers like _lists_ and _dicts_ always use the result of \__repr__ to represent the objects they contain.

__Recommendation__: always have at least a \__repr__ method. If there is no \__str__ method, it'll fall back on \__repr__ method.

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

myCar = Car('green', 100)
```

```python
print(myCar)
```

```python
myCar
```

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

    def __str__(self):
        return f'a {self.color} car'

    def __repr__(self):
        return f'a {self.color} car with {self.mileage} mileage'

myCar = Car('green', 100)
```

```python
print(myCar)
```

```python
myCar
```

```python
str(myCar)
```

```python
str([myCar])
```

```python
'{}'.format(myCar)
```

```python
repr(myCar)
```

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

    def __repr__(self):
        return f'a {self.color!r} {self.__class__.__name__} with {self.mileage!r} mileage'

myCar = Car('green', 100)
```

```python
str(myCar)
```

```python
myCar
```

## 4.3 Defining Your Own Exception Classes 

By defining your own error types, the potential error cases can stand out clearly, and as a result, the functions and modules will become more maintainable. The custom eror types can also provide additional debugging information.

```python
class BaseValidationError(ValueError):
    pass

class NameTooShortError(BaseValidationError):
    pass

class NameTooLongError(BaseValidationError):
    pass

def validate(name):
    if len(name)<3:
        raise NameTooShortError(name)
    if len(name)>20:
        raise NameTooLongError(name)
```

```python
validate('ly')
```

## 4.4 Cloning Objects for Fun and Profit

- Assignment: only bind names to an object
- shallow copy: constructs a new compound object and then insert _references_ into it to the objects found in the original;
- deep copy: constructs a new compound object and then recursively, inserts _copies_ into it of the objects found in the orignal.

Python documentation: https://docs.python.org/3/library/copy.html

copy module has copy() and deepcopy() to perform shallow and deep copies. This can also be used to arbitrary objects like custom classes. 

```python
# assignment: only bind names to an object

a = [[1,2.3], [4,7,3], [2,5,6]]
b = a
b.append([10, 11])
print(f'a is {a}')
print(f'b is {b}')
```

```python
# shallow copy: only one level

a = [[1,2.3], [4,7,3], [2,5,6]]
b = a.copy()
b.append([10, 11])
print(f'a is {a}')
print(f'b is {b}')
```

```python
a = [[1,2.3], [4,7,3], [2,5,6]]
b = a.copy()
b[2][1] = 4.88
print(f'a is {a}')
print(f'b is {b}')
```

```python
a = [[1,2.3], [4,7,3], [2,5,6]]
b = list(a)
b[2] = [4.88, 3]
print(f'a is {a}')
print(f'b is {b}')
```

```python
# list: create shallow copy, i.e. object changed but child object not changed. 
# But we can still change the reference to the child object. 
a = [[1,2.3], [4,7,3], [2,5,6]]
b = list(a)
b[2][1] = 1.88
print(f'a is {a}')
print(f'b is {b}')
```

```python
# Create deep copy

import copy

a = [[1,2.3], [4,7,3], [2,5,6]]
b = copy.deepcopy(a)
b[2][1] = 1.88
print(f'a is {a}')
print(f'b is {b}')
```

## 4.5 Abstract Base Classes Keep Inheritance in Check

ABS ensures that derived classes implement particular methods from the base class. 

```python

# base class
class Base():
    def foo(self):
        raise NotImplementedError()

    def bar(self):
        raise NotImplementedError()


class Concrete(Base):
    def foo(self):
        return 'foo() called'
```

```python
obj = Base()
obj.foo()
```

The base object can be instantiated!!! BUT we don't want that. 

```python
obj = Concrete()
obj.foo()
```

Also the subclass can be defined EVEN THOUGH bar() is not defined. We ALSO don't want that!!!

```python
# with abc class: 
from abc import ABCMeta, abstractmethod

class Base(metaclass = ABCMeta):
    @abstractmethod
    def foo(self):
        pass

    @abstractmethod
    def bar(self):
        pass


class Concrete(Base):
    def foo(self):
        pass
```

```python
# base class cannot be instantiated
baseObj = Base()
```

```python
# subclass cannot be defined when method is not defined
concreteObj = Concrete()
```

```python
# with abc class: subclass cannot be defined when method is not defined
from abc import ABCMeta, abstractmethod

class Base(metaclass = ABCMeta):
    @abstractmethod
    def foo(self):
        pass

    @abstractmethod
    def bar(self):
        pass


class Concrete(Base):
    def foo(self):
        pass

    def bar(self):
        pass
```

```python
concreteObj = Concrete()
```

```python
concreteObj.foo()
```

## 4.6 Namedtuples

Allows referencing contents by string. Tuples are immutable, i.e. can't change the content after creation. 

Namedtuples are a memory-efficient shortcut to defining an immutable class in Python manually. It can also help to clean up the code by enforcing an easier-to-understand structure on data.

Built-in Helper Methods: starts with single underscore prefix, but they're part of the public interface and it's ok to use them.
- \_asdict(): returns the contents of a namedtuple as a dictionary
- \_replace(): creates a shallow copy of a tuple and allows you to selectively replace some of its fields
- \_make(): can be used to create new instances of namedtuple from a sequence or iterable

```python
# subclassing namedtuples: can add other calss and add methods and new properties
Car = namedtuple('Car', 'color mileage')

class CarWithMethods(Car):
    def hexcolor(self):
        if self.color == 'red':
            return '#ff0000'
        else:
            return '#000000'
```

```python
myCar = CarWithMethods('red', 100)
```

```python
myCar.color
```

```python
myCar.hexcolor()
```

```python
# to add a new field to namedtuple is tricky, easiest way is to use _fields property

Car = namedtuple('Car', 'color mileage')
ElectricCar = namedtuple('ElectricCar', Car._fields + ('charge',))
```

```python
ElectricCar('grey', 3000, 10)
```

## 4.7 Class VS Instance Variable Pitfalls

- class variable: for data shared by all instances of a class. They belong to a class, not a specific instance and are shared among all the instances of a class.
- Instance variables: for data that is unique to each instance.
- Class variables can be "shadowed" by instance variables in the same name, so it's easy to accidentally override class variables in a way that introduces bugs and odd behaviour.

```python
class Dog:
    num_legs = 4 # This is a class variable

    def __init__(self, name):
        self.name = name # name is an instance variable
```

```python
myDog = Dog('Jimmy')
hisDog = Dog('Sally')
```

```python
myDog.__class__.num_legs, hisDog.num_legs
```

```python
myDog.num_legs=5 #!!! We accidentally created instance variable for myDog
```

```python
myDog.num_legs, myDog.__class__.num_legs, hisDog.num_legs
```

```python
myDog.__class__.num_legs = 6 # This changes the class variable
```

```python
myDog.num_legs, myDog.__class__.num_legs, hisDog.num_legs
```

## 4.8 Instance, Class, and Static Methods 

- instance method: need a class instance, can access the instance through _self_.
- class method: don't need a class instance. They can't access the instance (self) but they have access to the class itself via cls.
- static method: don't have access to cls or self. They work like regular functions but belong to the class' namespace. 


```python
class myClass():
    def __init__(self): 
        pass

    def instanceMethod(sef):
        return "instance method calling", sef

    @classmethod
    def classMethod(classname):
        return "class method calling", classname

    @staticmethod
    def staticMethod(x, y):
        return x+y
```

```python
myclass = myClass()
myclass.instanceMethod()
```

```python
myclass.classMethod()
```

```python
myclass.staticMethod(1,2)
```

```python
# Pizza example

import math

class Pizza:
    def __init__(self, radius, ingredients):
        self.radius = radius
        self.ingredients = ingredients

    def __repr__(self):
        return (f'Pizza {self.radius!r}, '
                f'{self.ingredients!r}')

    def area(self):
        return self.circle_area(self.radius)

    @staticmethod
    def circle_area(r):
        return r**2 * math.pi

    @classmethod
    def margherita(cls):
        return cls(6, ['mozzarella', 'tomato'])

    @classmethod
    def prosciutto(cls):
        return cls(6, ['mozzarella', 'tomato', 'ham'])
```

```python
myFavorite = Pizza.margherita()
```

```python
myFavorite
```

```python
myFavorite.area()
```
